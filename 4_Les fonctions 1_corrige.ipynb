### Structure élémentaire d'une fonction

#### Une fonction sans argument, qui ne renvoi rien

Ce programme affiche les initiales après avoir demandé à l'utilisateur son nom et son prénom.

```python
prenom = input("Quel est ton prénom ? ")
nom = input("Quel est ton nom ? ")
print("Tes initiales sont",prenom[0]+nom[0])
```

Ecrivez une fonction qui réalise la même chose. La fonction affiche le résultat mais ne renvoi aucune donnée :

```python
def initiales():
    prenom = input("Quel est ton prénom ? ")
    nom = input("Quel est ton nom ? ")
    print("Tes initiales sont",prenom[0]+nom[0])
```

Testez cette fonction ci-dessous :

```python
initiales()
```

#### Une fonction sans argument qui renvoi une chaine

Ecrivez une fonction qui demande toujours un nom et un prénom mais n'affiche plus rien en console et retourne une chaine contenant les initiales :

```python
def initiales():
    prenom = input("Quel est ton prénom ? ")
    nom = input("Quel est ton nom ? ")
    return (prenom[0] + nom[0]).upper()
```

Testez cette fonction ci-dessous :

```python
a = initiales()
print(a)
```

#### Une fonction recevant des arguments

Ecrivez une fonction qui reçoit en argument un nom et un prénom retourne une chaine contenant les initiales :

```python
def initiales(prenom, nom):
    return (prenom[0] + nom[0]).upper()
```

Testez cette fonction ci-dessous :

```python
a = initiales('Julien', 'marchand')
print(a)
```

### Des fonctions qui renvoient un booléen

Ecrivez une fonction **estPair(a)** qui renvoi True si a est pair et False si a est impair. l'instruction **else** est interdite :

```python
def estPair(a):
    if a%2:
        return False
    return True
```

Votre fonction doit permettre le fonctionnement du script suivant :

```python
if estPair(int(input("Ecrivez un entier "))):
    print("Ce nombre est pair")
else:
    print("Ce nombre est impair")
```

Ecrivez une fonction **estLeCarreDe(a, b)** qui renvoi True si a est le carré de b, False dans le cas contraire. l'instruction **else** est interdite :

```python
def estLeCarreDe(a, b):
    if a==b**2:
        return True
    return False
```

Votre fonction doit permettre le fonctionnement du script suivant :

```python
n1=float(input("Donner un 1er nombre  "))
n2=float(input("Donner un second nombre  "))
if estLeCarreDe(n1, n2):
    print(n1," est le carré de ", n2)
else:
    print(n1," n'est pas le carré de ", n2)
```

Ecrivez une fonction estBisextile(a) qui renvoi True si a est une année bisextile et False si elle ne l'est pas.

**Info :** Les années bissextiles sont celles qui sont divisibles par 4, sauf lorsqu’elles débutent un siècle non multiple de 400. (Par exemple : l’année 2000 était bissextile : 2000 débute un siècle mais 2000/400 = 50 ; l’année 2100 ne sera pas bissextile).

L'instruction **else** est interdite :

```python
def estBissextile(a):
    if a%400 == 0:
        return True 
    elif a%100 == 0:
        return False
    elif a%4 == 0:
        return True        
    return False

```

Proposez un script pour tester cette fonction

```python
if estBissextile(int(input("Rentre une année pour savoir si elle est bissextile : "))):
    print("Année Bissextile")
else:
    print("Année Non-Bissextile")
```
