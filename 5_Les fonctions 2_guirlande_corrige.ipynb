## Des dessins avec TURTLE

### Exemples d'utilisation de turtle

Turtle est une petite tortue que l'on peut déplacer librement sur l'écran pour la faire dessiner. il faut tout d'abord importer les fonctions de turtle dont nous avons besoin 

```python
from turtle import forward, backward, left, right, done
```

```python
forward(200) #avance de 200px
left(90) #tourne à gauche de 90°
forward(200)
right(45) #tourne à droite de 45°
backward(100) #recule de 100px
done() #commence à dessiner
```

On peut ajouter des couleurs !

```python
from turtle import fillcolor, pencolor, begin_fill, end_fill
```

```python
pencolor("blue")
fillcolor("orange")

begin_fill() #commence un tracé a remplir
forward(100)
left(90)
forward(100)
end_fill() #rempli ce tracé
done()
```

On peut se déplacer directement à une position en traçant ou sans tracer ! tracer des cercles ...

```python
from turtle import goto, up, down, circle
```

```python
up() #lève le crayon
goto(-200, 0)
down()
circle(100)
up()
goto(200, -100)
down()
circle(50)
goto(100,-100)
circle(50)
done()
```

On peut également demander à la tortue d'aller plus vite, voire beaucoup plus vite.


```python
from turtle import speed
```

```python
#la tortue commence a son rythme, tranquile ...
up()
goto(0, -200)
down()
circle(200)

speed('fastest') #accélération fulgurante !!!

for i in range(5, 200, 5):
    up()
    goto(0, -200 + i)
    down()
    circle(200 - i) 
    
done()
```

### Des fonctions avec Turtle

Nous allons progressivement réaliser le dessin d’une guirlande colorée. Pour cela, nous allons créer des fonctions élémentaires, et les imbriquer pour réaliser un dessin de plus en plus compliqué et arriver à ceci : 

#### Les formes de base

Proposez une fonction **carre(longCote)** qui dessine un carré de taille définie par le paramètre **longCote**. Avant de lancer cette fonction, le crayon est levé, il devra être levé avant de quitTer la fonction.

```python
def carre(longCote):
    down()
    for n in range(4):
        forward(longCote)
        right(90)
    up()
```

Le programme suivant doit tracer 3 carrés de taille différentes !

```python
carre(200)
goto(50,-50)
carre(100)
goto(75,-75)
carre(50)
done()
```

Proposez une fonction **triangle(longCote)** qui dessine un triangle équilatéral, pointe vers le bas,  de taille définie par le paramètre **longCote**. Avant de lancer cette fonction, le crayon est levé, il devra être levé avant de quitTer la fonction.

```python
def triangle(longCote):
    down()
    for n in range(3):
        forward(longCote)
        right(120)
    up()
```

Le programme suivant doit tracer 3 triangles de taille différentes et concentriques !

```python
triangle(200)
goto(50,-50)
triangle(100)
goto(75,-75)
triangle(50)
done()
```

#### Un peu de couleur

Ecrivez deux nouvelles fonctions **triangleJaune(longCote)** et **carreRouge(longCote)** qui dessinent deux figures colorées en utilisant obligatoirement les fonctions précédentes : **carre(longCote)** et **triangle(longCote)**

```python
def triangleJaune (longCote):
    fillcolor("yellow")
    begin_fill()
    triangle(longCote)
    end_fill()

def carreRouge (longCote):
    fillcolor("red")
    begin_fill()
    carre(longCote)
    end_fill()
```

Testez ces deux fonctions ci-dessous :

```python
carreRouge (200)
triangleJaune (100)
done()
```

#### Un motif de base pour la guirlande

Ecrivez une fonction **motif(longCote)** qui dessine le motif ci-dessous à partir des deux fonctions précédentes. Notez la position de la tortue après exécution du motif. 

```python
def motif (longCote):
    carreRouge(longCote)
    triangleJaune(longCote)
    forward(longCote)
```

Testez ces deux fonctions ci-dessous :

```python
motif(200)
done()
```

#### La guirlande complète

Pour dessiner une guirlande complète, il faut répéter le motif neuf fois de suite avec un décalage de quelques degrès (10 par exemple) et en resectant les tailles suivantes pour chacun des motifs : 25, 50, 75, 100, 125, 100, 75, 50, 25.   
Proposez une fonction **guirlande()** qui dessine la guirlande à partir de la potition (-300, 100).

```python
def guirlande():
    up()    
    goto(-300, 100)
    right(40)
    taille = (25, 50, 75, 100, 125, 100, 75, 50, 25)
    #dessine la guirlande
    for t in taille:
        motif(t)
        left(10)
    right(50)
  
```

La cellule suivante permet de tester votre guirlande.

```python
speed("fastest")
guirlande()
done()
```

La guirlande précédente est assez figée : 25px entre deux motifs consécutif et une position fixe. On souhaite rendre notre guirlande paramétrable : proposez une fonction **guirlande(x0, y0, delta)** qui permette de choisir sa position de départ et l'écart en pixel entre deux motifs consécutifs.

```python
def guirlande(x0, y0, delta):
    up()    
    goto(x0,y0)
    right(40)
    #dessine la guirlande
    for x in range(1, 5):
        motif(x*delta)
        left(10)
    for x in range(5, 0, -1):
        motif(x*delta)
        left(10)
    right(50)
```

La cellule suivante permet de tester votre guirlande.

```python
speed("fastest")
#la même que précédemment
guirlande(-300, 100, 25)
#ou une autre
guirlande(-200, 150, 10)
guirlande(100, 150, 5)
done()
```

### Une autre idée de motif

```python
def carreAngleOrange(longCote):
    pencolor("blue")
    fillcolor("orange")
    down()
    for i in range(4):
        begin_fill()
        forward(longCote//2)
        right(90)
        forward(longCote//2)
        end_fill()
    up()

def bouleBleue(rayon):
    fillcolor("blue")
    right(90)
    forward(rayon)
    right(90)
    down()
    begin_fill()
    circle(rayon)
    end_fill()
    up()
    right(90)
    forward(rayon)
    right(90)    


carreAngleOrange(200)
bouleBleue(50)
done()
```

```python
def motif2 (longCote):

    forward(longCote//2)
    carreAngleOrange(longCote)
    bouleBleue(longCote//4)

    forward(longCote//2)
```

```python
up()
goto(-200, 0)
motif2(200)
left(20)
motif2(100)
left(20)
motif2(150)
done()
```

```python
def guirlande2(x0, y0, delta):
    up()    
    goto(x0,y0)
    right(40)
    #dessine la guirlande
    for x in range(1, 5):
        motif2(x*delta)
        left(10)
    for x in range(5, 0, -1):
        motif2(x*delta)
        left(10)
    right(50)
```

```python
speed("fast")
#la même que précédemment
guirlande2(-300, 100, 25)
#ou une autre
guirlande2(-200, 150, 10)
guirlande2(100, 150, 5)
done()
```
