### SOMME DE N TERMES

#### version itérative

```python
def sommeI(n):
    s = 0
    for i in range(1, n+1):
        s = s + i
    return s
resultat = sommeI(5)

tutor()
```

#### version récursive

```python
def sommeR(n):
    if n == 1:
        return(1)
    else:
        return n + sommeR(n - 1)
resultat = sommeR(5)

tutor()
```

### CALCUL D’UNE SUITE PAR RÉCURSIVITÉ

```python
def suite(n):
    """calcul par récursivité
    des éléments de la suite
    un+1 = un * 2 + 3
    u0=2"""
    
    if n == 0:
        return 2
    else:
        return 2 * suite(n-1) + 3
    
resultat = suite(5)

tutor()
```

### FIBONACCI

### version récursive

```python
def fiboR (n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fiboR(n-1) + fiboR(n-2)
resultat = fiboR(5)

tutor()
```

### version dynamique

```python
dicFibo = {0:0, 1:1}

def fiboD (n):
    global dicFibo
    if n in dicFibo:
        return dicFibo[n]
    else:
        r = fiboD(n-1) + fiboD(n-2)
        dicFibo[n] = r
        return r
resultat = fiboD (5)
tutor()
```
