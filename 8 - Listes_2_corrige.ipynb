Ecrivez une fonction **moyenne(lst)** qui reçoit une liste en argument et retourne la moyenne des termes de cette liste.  
moyenne([1, 2, 3, 4, 5]) renvoi 3.0

````python
def moyenne(lst):
    somme = 0
    for u in lst:
        somme = somme + u
    return somme / len(lst)
moyenne([1, 2, 3, 4, 5]) 
````

Ecrivez une fonction **minimum(lst)** qui prend une liste en argument et renvoi la plus petite valeur de cette liste sans la modifier.

````python
def minimum(lst):
    mini = lst[0]
    for u in lst:
        if u < mini:
            mini = u
    return mini
minimum([12, 4, 5, 1, 8, 2, 9])
````

Ecrivez une fonction **carre(n)** qui renvoi une liste des carrés des n premiers entiers. 

````python
def carre(n):
    lst = []
    for u in range(1, n + 1):
        lst.append(u**2)
    return lst
carre(10)        
````

#### Avec une liste en argument

````python
#un exemple
def double(lst):
    lst2 = []
    for u in lst:
        lst2.append(2*u)
    return lst2

double([1, 5, 14, 100, 99])
````
