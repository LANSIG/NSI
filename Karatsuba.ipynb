````python
def add(lA, lB):
    """
    Addition des deux nombres lA et lB sous forme de liste de chiffre, poid faible en tete.
    lA et lB ne sont pas modifiÃ©es par la fonction
    retourne le rÃ©sultat sous forme d'une liste
    
    Tests automatiques
    >>> a = [2, 5]
    >>> b = [9, 2]
    >>> add(a,b)
    [1, 8]
    >>> a = [8, 9, 9, 9]
    >>> b = [1]
    >>> add(a,b)
    [9, 9, 9, 9]
    >>> a
    [8, 9, 9, 9]
    >>> b
    [1]
    >>> a = [9, 9, 9, 9]
    >>> b = [1]
    >>> add(a,b)
    [0, 0, 0, 0, 1]
    >>> a
    [9, 9, 9, 9]
    >>> b
    [1]
    >>> a = [5, 4, 3, 2, 1]
    >>> b = [5, 6, 7, 8, 9]
    >>> add(a,b)
    [0, 1, 1, 1, 1, 1]
    >>> a
    [5, 4, 3, 2, 1]
    >>> b
    [5, 6, 7, 8, 9]
    """
    #copie les listes pour ne pas les mofifier
    l1 = lA[:]
    l2 = lB[:]
    #Ã©galise la longueur des deux listes
    memeLen(l1, l2)
    
    r = 0 #retenue
    res = [] #rÃ©sultat
    for i in range(len(l1)):
       result =  l1[i] + l2[i] + r
       res.append(result % 10) #le chiffre des unitÃ©s au rÃ©sultat (poids fort)
       r = result // 10 #le chiffre des dizaine en retenue pour le niveau suivant

        
    return res
````

## Comparaison avec la multiplication naÃ¯ve

Voici la fonction **mulNaive(lst1, lst2)** qui rÃ©alise la multiplication des deux listes de chiffre avec lâ€™algorithme naÃ¯f (voir cours).

````python
def mulNaive(lA, lB):
    """
    Multiplication naive (comme a l'Ã©cole primaire) entre les deux listes lA et lB.
    Renvoi le rÃ©sultat et ne modifie pas les les listes lA et lB
    
    """
    l1 = lA[:]
    l2 = lB[:]
    
    res=[]#matrice des multiplications chiffre par chiffre
    for i in range(len(l1)):
        #pour chaque chiffre de l1 : l1[i]
        #que l'on multiplie par l2
        #le rÃ©sultat est une liste qu'on ajoute dans res
        r = 0 #retenue nulle au dÃ©but
        res.append([0]*i) #on dÃ©cale d'un cran a chaque chiffre de l1, on met des points au primaire ...
        for j in range(len(l2)):
            result = l1[i] * l2[j] + r
            res[i].append(result % 10) #le chiffre des unitÃ© va au rÃ©sultat
            r = result // 10 #celui des dizaine part en retenue
        if r != 0:
            #si il reste une reteune a la fin, on l'ajoute. Ã§a fera un chiffre de plus en poids fort
            res[i].append(r)
    
    result = []#resultat final
    for i in range(len(res)):
        result = add(result, res[i])
        
    return result
````
