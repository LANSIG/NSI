## 3 - Des dessins avec turtle

### Module turtle

Turtle est une petite tortue que l'on peut déplacer librement sur l'écran pour la faire dessiner. il faut tout d'abord importer les fonctions de turtle dont nous avons besoin.

````python
from turtle import forward, backward, left, right, done, fillcolor, pencolor
from turtle import begin_fill, end_fill, goto, up, down, circle, speed
````

### Un nouveau motif  
On vous donne deux fonctions **carreAngleOrange(longCote)** et **bouleBleue(rayon)** qui tracent respectivement un carré aux angles oranges et une boule bleue, tracé en bleu. 
Exécutez-la pour tester.

````python
def carreAngleOrange(longCote):
    fillcolor("yellow")
    pencolor("red")
    down()
    for i in range(4):
        begin_fill()
        forward(longCote//2)
        right(90)
        forward(longCote//2)
        end_fill()
    up()

def bouleBleue(rayon):
    fillcolor("grey")
    down()
    right(90)
    forward(rayon)
    right(90)
    begin_fill()
    circle(rayon)
    end_fill()
    right(90)
    forward(rayon)
    right(90)
    up()

#Pour tester les deux fonctions
up()
goto(-200, 0)
carreAngleOrange(200)
goto(200, 100)
bouleBleue(100)
done()
````
