### Créer une liste  

Créer une liste de nombre 
    
```python

    ```    
    
Créer une liste vide 
    
```python

    ```

Si un seul élément dans la liste : même syntaxe
    

Une liste peut contenir tout type de donnée, y compris une autre liste 
    
```python

    ```    
    
L'instruction **list()** permet de convertir tous les itérables en liste
    
```python
>>> lst1 = list(range(8))
>>> lst1    
[0, 1, 2, 3, 4, 5, 6, 7]
>>> lst2 = list("Hello")
>>> lst2    
['H', 'e', 'l', 'l', 'o']    
    ```     
   
Multiplier une liste et un entier permet de reproduire la liste :
    
```python
>>> lst = [1] * 5
>>> lst
[1, 1, 1, 1, 1]
>>> lst = [1, 2] * 5
>>> lst
[1, 2, 1, 2, 1, 2, 1, 2, 1, 2]   
    ```

Ecrivez une fonction **lstPair(n)** qui renvoi une liste des entiers pairs compris entre 1 et n.

````python
def lstPair(n):
    return list(range(2, n + 1, 2))

lstPair(100)
````

Ecrivez une fonction **lstPairImpair(n, pair)** qui renvoi une liste des entiers compris entre 1 et n. Cette liste sera constituée uniquement des entiers pairs si l'argument **pair** est True et uniquement des entiers impairs si l'argument **pair** est False.

````python
def lstPairImpair (n, paire = True):
    
    if paire == True:
        return list(range(2, n + 1, 2))
    else:
        return list(range(1, n + 1, 2))
        
print("Sans l'argument optionnel :\n", lstPairImpair(100), '\n')
print("Avec l'argument optionnel à True :\n", lstPairImpair(100, True), '\n')
print("Avec l'argument optionnel à False:\n", lstPairImpair(100, False), '\n')
````

### lire un élément d'une liste

Comme les chaines, les listes sont indexables; le premier élément porte l'index 0.
    
```python
>>> lst = [1, 2, 3, 4] 
>>> lst[0]
1
>>> lst[2]
3
    ```    
    
Dans les deux sens, le dernier éléments portant l'index -1. 
    
```python
>>> lst = [1, 2, 3, 4] 
>>> lst[-1]
3
>>> lst[-3]
2
    ```  

La fonction **len(liste)** retourne le nombre d'élément dans la liste
    
```python
>>> len(lst)
4
>>> len([0, 2, 4, 6, 8, 10])
6    
    ```    
    
Une liste peut être découpée, cette opération s'appelle le slicing.
    
```python
>>> lst = list(range(9))   
>>> lst[1:5]
[1, 2, 3, 4]
>>> lst[3:]
[3, 4, 5, 6, 7, 8]
>>> lst[:3]
[0, 1, 2]
    ```

### ajouter des éléments dans une liste

La méthode **append()** permet d'ajouter un élément à la fin de'une liste
    
```python
>>> lst = list(range(9))
>>> lst.append(12)
>>> lst.append(13)
>>> lst.append(14)
>>> lst
[0, 1, 2, 3, 4, 5, 6, 7, 8, 12, 13, 14]
    ```    
    
On peut également concaténer deux listes de cette façon :
    
```python
>>> lst = [1, 2, 3, 4] + [-1, -2, -3, -4]
>>> lst   
[1, 2, 3, 4, -1, -2, -3, -4]    
    ```

Ou en utilisant la méthode **extend()** de cette façon :
    
```python
>>> lst = [1, 2, 3, 4]
>>> lst.extend([-1, -2, -3, -4])
>>> lst   
[1, 2, 3, 4, -1, -2, -3, -4]    
    ```

Ecrivez une fonction **decoupStr(string)** qui reçoit en argument un mot et retourne une liste composée des lettres du mot sans doublon. par exemple decoupStr("coucou!!!") renvoi ['c', 'o', 'u', '!']. Le test **in** que vous avez déjà utilisé fonctionne sur les listes.

````python
def decoupStr(string):
    lst = []
    for u in string:
        if u not in lst:
            lst.append(u)
    return lst

decoupStr("coucou!!!")
````

### retirer des éléments dans une liste

La méthode **pop()** permet de retirer le dernier élément de la liste et le **retourner**. 
    
```python
>>> lst = list(range(2, 8))
>>> lst    
[2, 3, 4, 5, 6, 7]
>>> x = lst.pop()
>>> x
7
>>> lst
[2, 3, 4, 5, 6]    
    ```    
    
Ou n'importe quel élément par son indice :

    
```python
>>> x = lst.pop(2)
>>> x
4
>>> lst
[2, 3, 5, 6]    
    ``` 

La méthode remove supprime la première occurence d'une valeur sans la retourner. Attention, si cette valeur n'est pas dans la liste, cela déclenche une exception.
    
```python
>>> lst = [1, 2, 3, 5, 4, 5, 5, 6, 7, 8]
>>> lst.remove(5)
>>> lst
[1, 2, 3, 4, 5, 5, 6, 7, 8]    
>>> lst.remove(5)
>>> lst
[1, 2, 3, 4, 5, 6, 7, 8]    
>>> lst.remove(12)
ValueError: list.remove(x): x not in list

    ```
