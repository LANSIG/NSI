### Des fonctions et des boucles avec turtle

#### des escaliers ...

écrivez deux fonctions **marcheHaut()** et **marcheBas()** qui dessinent respectivement les deux figures ci-dessous. La hauteur et la longueur de la marche font 20 pixels : 

```python
def marcheHaut ():
    down()
    left(90)
    forward(15)
    right(90)
    forward(15)
    up()

def marcheBas ():
    down()
    right(90)
    forward(15)
    left(90)
    forward(15)
    up()
    
marcheHaut()
marcheBas()
done()
```

Ecrivez deux fonctions **escalierMontant(nbMarche)** et **escalierDescendant(nbMarche)** qui réalisent respectivement les deux figures ci-dessous pour nbMarche = 5 : 

```python
def escalierMontant (nbMarche):
    for i in range(nbMarche):
        marcheHaut ()

def escalierDescendant (nbMarche):
    for i in range(nbMarche):
        marcheBas ()

escalierMontant(5)
escalierDescendant(5)
done()
```

En utilisant les fonctions écrites précédemment, réalisez la figure suivante : 

```python
up()
backward(300)
down()
for i in range (3):
    escalierMontant(6)
    escalierDescendant(6)
done()
```

Idem avec la figure ci-dessous. Il y a 6 escaliers, le premier a 1 marches, les suivants ont une marche de plus que le précédent : 

```python
up()
backward(310)
down()
for i in range (6):
    escalierMontant(i + 1)
    escalierDescendant(i + 1)
done()
```

### Ciel étoilé

Vous aurez besoin ici de quelques fonctions turtle supplémentaires.

```python
from turtle import bgcolor, goto, fillcolor, begin_fill, end_fill, pencolor
#les dimensions de la fenetre de dessin turtle sont
largeurFenetre = 315*2
hauteurFenetre = 240*2
```

écrivez une fonction **etoile(long)** qui trace avec Turtle, une étoile composée de cinq segments de longueur **long**, remplie en blanc, avec les traits blancs. La fonction termine en levant le crayon.  
***Aide :*** la somme des angles des cinq branches de l’étoile vaut 180° 

```python
def etoile (long):
    fillcolor('white')
    pencolor('white')
    down()
    begin_fill()
    for i in range (5):
       forward(long)
       left(144)   #180-(180/5)
    end_fill()
    up()
```

Testez ci-dessous sur un fond noir :

```python
bgcolor('black')
etoile(200)
done()
```

Ecrivez une fonction **etoileAlea(x,y)** qui utilise la fonction etoile(long). Elle place une étoile au point de coordonnées (x, y) dont l’orientation et la taille sont aléatoires. La taille variera entre 10 et 50 pixels. Exemple :  
***Aide :*** Pour obtenir un nombre aléatoire compris entre i et j, vous pouvez utiliser la fonction randint(i, j) qui se trouve dans le module random.

```python
from random import randint

def etoileAlea (x,y):
    goto(x,y)
    left(randint(0,360))
    etoile(randint(10,50))
```

Testez ci-dessous sur un fond noir :

```python
bgcolor('black')
etoileAlea(0, 0)
etoileAlea(100, 0)
etoileAlea(0, 100)
etoileAlea(100, 100)
done()
```

Ecrivez le programme permettant de répartir aléatoirement sur l’ensemble de la fenêtre graphique une centaine de ces petites étoiles sur un fond bleu :

```python
speed('fastest')
up()
bgcolor('blue')

for i in range(100):
    xa = randint(- largeurFenetre // 2, largeurFenetre // 2)
    ya = randint(- hauteurFenetre // 2, hauteurFenetre // 2)
    etoileAlea(xa,ya)
done()
```
